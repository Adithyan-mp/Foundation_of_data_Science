```python
import numpy as np
import matplotlib.pyplot as plt

def random_walk_1d(steps):
    path = np.zeros(steps)
    for i in range(1, steps):
        step = np.random.choice([-1, 1])
        path[i] = path[i - 1] + step
    return path

num_steps = int(input("Enter the number of steps: "))
walk = random_walk_1d(num_steps)

plt.subplot(2, 1, 1)
plt.scatter(range(num_steps), walk)
plt.title("1D Random Walk")
plt.xlabel("Steps")
plt.ylabel("Position")

plt.subplot(2, 1, 2)
plt.hist(walk, bins=int(max(walk) - min(walk)) + 1, color='blue', alpha=0.7, rwidth=0.8)
plt.title("Histogram with X as Position and Y as Frequency")
plt.xlabel("Position")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
def prob(x, N):
    count = 0
    for i in range(N):
        if walk[i] == x:
            count += 1
    return count / N

print(prob(0, 100))
```

```python
def prob_formula(x, N):
    probability = 0
    if x == 0:
        probability = (1/2) ** N
    elif x % 2 == 0 and N % 2 == 0:
        probability = ((1/2) ** N) * (np.math.factorial(N) // (np.math.factorial((N + x) // 2) * np.math.factorial((N - x) // 2)))
    elif x % 2 != 0 and N % 2 != 0:
        probability = ((1/2) ** N) * (np.math.factorial(N) // (np.math.factorial(N + x // 2) * np.math.factorial(N - x // 2)))
    else:
        return 0
    return probability

print(prob_formula(2, 100))
```
